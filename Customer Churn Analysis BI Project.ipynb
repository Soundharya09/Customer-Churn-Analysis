# CUSTOMER CHURN ANALYSIS - POWER BI PROJECT 

```python
import pandas as pd
```

```python
train = pd.read_csv('customer_churn_dataset-training.csv')
test = pd.read_csv('customer_churn_dataset-testing.csv')

df = pd.concat([train, test], ignore_index = True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
numeric_cols = ["Age", "Tenure", "Usage Frequency", "Support Calls",
    "Payment Delay", "Total Spend", "Last Interaction", "Churn"]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors = "coerce")
```

```python
df.head()
```

```python
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
df["Gender"] = df["Gender"].fillna("Unknown")
```

```python
df["CustomerID"] = (
    pd.to_numeric(df["CustomerID"], errors="coerce").astype("Int64").astype(str))
```

```python
df.to_csv("customer_churn_cleaned.csv", index=False)
```
